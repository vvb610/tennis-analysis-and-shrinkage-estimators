# Functions

```python
def compute_matchup_stats_Oracle(match):
    s_itt = match["p1 tourney serve adv Oracle"]
    r_itt = match["p1 tourney return adv Oracle"]
    
    s_jtt = match["p2 tourney serve adv Oracle"]
    r_jtt = match["p2 tourney return adv Oracle"]
    
    sg_itt = match["p1 tourney service game adv Oracle"]
    rg_itt = match["p1 tourney return game adv Oracle"]
    
    sg_jtt = match["p2 tourney service game adv Oracle"]
    rg_jtt = match["p2 tourney return game adv Oracle"]
    
    sets_itt = match["p1 tourney set adv Oracle"]
    sets_jtt = match["p2 tourney set adv Oracle"]
    
    t_itt = match["p1 tourney tiebreak adv Oracle"]
    t_jtt = match["p2 tourney tiebreak adv Oracle"]
    
    # Pr[i wins serve on j at tournament t on day t]
    s_ijtt = (s_itt * (1 - r_jtt)) / ((s_itt * (1 - r_jtt)) + (r_jtt * (1 - s_itt)))
    # Pr[j wins serve on i at tournament t on day t]
    s_jitt = (s_jtt * (1 - r_itt)) / ((s_jtt * (1 - r_itt)) + (r_itt * (1 - s_jtt)))
    
    # Pr[i wins service game on j at tournament t on day t]
    sg_ijtt = (sg_itt * (1 - rg_jtt)) / ((sg_itt * (1 - rg_jtt)) + (rg_jtt * (1 - sg_itt)))
    # Pr[j wins service game on i at tournament t on day t]
    sg_jitt = (sg_jtt * (1 - rg_itt)) / ((sg_jtt * (1 - rg_itt)) + (rg_itt * (1 - sg_jtt)))
    
    # Pr[i wins set on j at tournament t on day t]
    sets_ijtt = (sets_itt * (1 - sets_jtt)) / ((sets_itt * (1 - sets_jtt)) + (sets_jtt * (1 - sets_itt)))
    
    # Pr[i wins tiebreak on j at tournament t on day t]
    t_ijtt = (t_itt * (1 - t_jtt)) / ((t_itt * (1 - t_jtt)) + (t_jtt * (1 - t_itt)))
    
    
    return s_ijtt, sg_ijtt, sets_ijtt, t_ijtt, s_jitt, sg_jitt
```

### Compute Oracle Estimator coef $\frac{v^2}{v^2 + (T - \sigma)^2}$ using clusted variance by match

```python
matches["p1 oracle coef serve"] = matches["p1 serve win rate variance"] / (matches["p1 serve win rate variance"] + (matches["all player year serve win rate"] - matches["p1 serve win rate"])**2)
matches["p1 oracle coef return"] = matches["p1 return win rate variance"] / (matches["p1 return win rate variance"] + ((1 - matches["all player year serve win rate"]) - matches["p1 return win rate"])**2)
matches["p1 oracle coef service game"] = matches["p1 service game win rate variance"] / (matches["p1 service game win rate variance"] + (matches["all player year service game win rate"] - matches["p1 service game win rate"])**2)
matches["p1 oracle coef return game"] = matches["p1 return game win rate variance"] / (matches["p1 return game win rate variance"] + ((1 - matches["all player year service game win rate"]) - matches["p1 return game win rate"])**2)
matches["p1 oracle coef set"] = matches["p1 set win rate variance"] / (matches["p1 set win rate variance"] + (0.5 - matches["p1 set win rate"])**2)
matches["p1 oracle coef tiebreak"] = matches["p1 tiebreak win rate variance"] / (matches["p1 tiebreak win rate variance"] + (0.5 - matches["p1 tiebreak win rate"])**2)

matches["p2 oracle coef serve"] = matches["p2 serve win rate variance"] / (matches["p2 serve win rate variance"] + (matches["all player year serve win rate"] - matches["p2 serve win rate"])**2)
matches["p2 oracle coef return"] = matches["p2 return win rate variance"] / (matches["p2 return win rate variance"] + ((1 - matches["all player year serve win rate"]) - matches["p2 return win rate"])**2)
matches["p2 oracle coef service game"] = matches["p2 service game win rate variance"] / (matches["p2 service game win rate variance"] + (matches["all player year service game win rate"] - matches["p2 service game win rate"])**2)
matches["p2 oracle coef return game"] = matches["p2 return game win rate variance"] / (matches["p2 return game win rate variance"] + ((1 - matches["all player year service game win rate"]) - matches["p2 return game win rate"])**2)
matches["p2 oracle coef set"] = matches["p2 set win rate variance"] / (matches["p2 set win rate variance"] + (0.5 - matches["p2 set win rate"])**2)
matches["p2 oracle coef tiebreak"] = matches["p2 tiebreak win rate variance"] / (matches["p2 tiebreak win rate variance"] + (0.5 - matches["p2 tiebreak win rate"])**2)
```

### Compute Oracle Estimator coef $\frac{v^2}{v^2 + (T - \sigma)^2}$ using pooled variance dividing by number of attempts

```python
matches["p1 oracle coef serve2"] = matches["p1 serve win rate variance2"] / (matches["p1 serve win rate variance2"] + (matches["all player year serve win rate"] - matches["p1 serve win rate"])**2)
matches["p1 oracle coef return2"] = matches["p1 return win rate variance2"] / (matches["p1 return win rate variance2"] + ((1 - matches["all player year serve win rate"]) - matches["p1 return win rate"])**2)
matches["p1 oracle coef service game2"] = matches["p1 service game win rate variance2"] / (matches["p1 service game win rate variance2"] + (matches["all player year service game win rate"] - matches["p1 service game win rate"])**2)
matches["p1 oracle coef return game2"] = matches["p1 return game win rate variance2"] / (matches["p1 return game win rate variance2"] + ((1 - matches["all player year service game win rate"]) - matches["p1 return game win rate"])**2)
matches["p1 oracle coef set2"] = matches["p1 set win rate variance2"] / (matches["p1 set win rate variance2"] + (0.5 - matches["p1 set win rate"])**2)
matches["p1 oracle coef tiebreak2"] = matches["p1 tiebreak win rate variance2"] / (matches["p1 tiebreak win rate variance2"] + (0.5 - matches["p1 tiebreak win rate"])**2)

matches["p2 oracle coef serve2"] = matches["p2 serve win rate variance2"] / (matches["p2 serve win rate variance2"] + (matches["all player year serve win rate"] - matches["p2 serve win rate"])**2)
matches["p2 oracle coef return2"] = matches["p2 return win rate variance2"] / (matches["p2 return win rate variance2"] + ((1 - matches["all player year serve win rate"]) - matches["p2 return win rate"])**2)
matches["p2 oracle coef service game2"] = matches["p2 service game win rate variance2"] / (matches["p2 service game win rate variance2"] + (matches["all player year service game win rate"] - matches["p2 service game win rate"])**2)
matches["p2 oracle coef return game2"] = matches["p2 return game win rate variance2"] / (matches["p2 return game win rate variance2"] + ((1 - matches["all player year service game win rate"]) - matches["p2 return game win rate"])**2)
matches["p2 oracle coef set2"] = matches["p2 set win rate variance2"] / (matches["p2 set win rate variance2"] + (0.5 - matches["p2 set win rate"])**2)
matches["p2 oracle coef tiebreak2"] = matches["p2 tiebreak win rate variance2"] / (matches["p2 tiebreak win rate variance2"] + (0.5 - matches["p2 tiebreak win rate"])**2)
```

### Compute Oracle Shrunken rates using coefs1

```python
matches["p1 serve win rate Oracle"] = matches["p1 serve win rate"] - matches["p1 oracle coef serve"] * (matches["p1 serve win rate"] - matches["all player year serve win rate"])
matches["p1 return win rate Oracle"] = matches["p1 return win rate"] - matches["p1 oracle coef return"] * (matches["p1 return win rate"] - (1 - matches["all player year serve win rate"]))
matches["p1 service game win rate Oracle"] = matches["p1 service game win rate"] - matches["p1 oracle coef service game"] * (matches["p1 service game win rate"] - matches["all player year service game win rate"])
matches["p1 return game win rate Oracle"] = matches["p1 return game win rate"] - matches["p1 oracle coef return game"] * (matches["p1 return game win rate"] - (1 - matches["all player year service game win rate"]))
matches["p1 set win rate Oracle"] = matches["p1 set win rate"] - matches["p1 oracle coef set"] * (matches["p1 set win rate"] - 0.5)
matches["p1 tiebreak win rate Oracle"] = matches["p1 tiebreak win rate"] - matches["p1 oracle coef tiebreak"] * (matches["p1 tiebreak win rate"] - 0.5)

matches["p2 serve win rate Oracle"] = matches["p2 serve win rate"] - matches["p2 oracle coef serve"] * (matches["p2 serve win rate"] - matches["all player year serve win rate"])
matches["p2 return win rate Oracle"] = matches["p2 return win rate"] - matches["p2 oracle coef return"] * (matches["p2 return win rate"] - (1 - matches["all player year serve win rate"]))
matches["p2 service game win rate Oracle"] = matches["p2 service game win rate"] - matches["p2 oracle coef service game"] * (matches["p2 service game win rate"] - matches["all player year service game win rate"])
matches["p2 return game win rate Oracle"] = matches["p2 return game win rate"] - matches["p2 oracle coef return game"] * (matches["p2 return game win rate"] - (1 - matches["all player year service game win rate"]))
matches["p2 set win rate Oracle"] = matches["p2 set win rate"] - matches["p2 oracle coef set"] * (matches["p2 set win rate"] - 0.5)
matches["p2 tiebreak win rate Oracle"] = matches["p2 tiebreak win rate"] - matches["p2 oracle coef tiebreak"] * (matches["p2 tiebreak win rate"] - 0.5)
```

```python
# calculate no sos adjust p1 serve_tt, pl_service_game_tt, etc.
matches["p1 serve_tt Oracle no sos adjust"] = (matches["p1 serve win rate Oracle"] * (1 - matches["p2 return win rate Oracle"])) / ((matches["p1 serve win rate Oracle"] * (1 - matches["p2 return win rate Oracle"])) + (matches["p2 return win rate Oracle"] * (1 - matches["p1 serve win rate Oracle"])))
matches["p2 serve_tt Oracle no sos adjust"] = (matches["p2 serve win rate Oracle"] * (1 - matches["p1 return win rate Oracle"])) / ((matches["p2 serve win rate Oracle"] * (1 - matches["p1 return win rate Oracle"])) + (matches["p1 return win rate Oracle"] * (1 - matches["p2 serve win rate Oracle"])))

matches["p1 service_game_tt Oracle no sos adjust"] = (matches["p1 service game win rate Oracle"] * (1 - matches["p2 return game win rate Oracle"])) / ((matches["p1 service game win rate Oracle"] * (1 - matches["p2 return game win rate Oracle"])) + (matches["p2 return game win rate Oracle"] * (1 - matches["p1 service game win rate Oracle"])))
matches["p2 service_game_tt Oracle no sos adjust"] = (matches["p2 service game win rate Oracle"] * (1 - matches["p1 return game win rate Oracle"])) / ((matches["p2 service game win rate Oracle"] * (1 - matches["p1 return game win rate Oracle"])) + (matches["p1 return game win rate Oracle"] * (1 - matches["p2 service game win rate Oracle"])))

matches["p1 tiebreak_tt Oracle no sos adjust"] = (matches["p1 tiebreak win rate Oracle"] * (1 - matches["p2 tiebreak win rate Oracle"])) / ((matches["p1 tiebreak win rate Oracle"] * (1 - matches["p2 tiebreak win rate Oracle"])) + (matches["p2 tiebreak win rate Oracle"] * (1 - matches["p1 tiebreak win rate Oracle"])))

matches["p1 set_tt Oracle no sos adjust"] = (matches["p1 set win rate Oracle"] * (1 - matches["p2 set win rate Oracle"])) / ((matches["p1 set win rate Oracle"] * (1 - matches["p2 set win rate Oracle"])) + (matches["p2 set win rate Oracle"] * (1 - matches["p1 set win rate Oracle"])))


```

```python
# Generate relative abilities JS, we use unshrunken opp stats (allows a possibility for improvement)
matches["p1 relative serve Oracle"] = matches["p1 serve win rate Oracle"] - (1 - matches["p1 opp return win rate"])
matches["p2 relative serve Oracle"] = matches["p2 serve win rate Oracle"] - (1 - matches["p2 opp return win rate"])

matches["p1 relative return Oracle"] = matches["p1 return win rate Oracle"] - (1 - matches["p1 opp serve win rate"])
matches["p2 relative return Oracle"] = matches["p2 return win rate Oracle"] - (1 - matches["p2 opp serve win rate"])

matches["p1 relative service games Oracle"] = matches["p1 service game win rate Oracle"] - (1 - matches["p1 opp return game win rate"])
matches["p2 relative service games Oracle"] = matches["p2 service game win rate Oracle"] - (1 - matches["p2 opp return game win rate"])

matches["p1 relative return games Oracle"] = matches["p1 return game win rate Oracle"] - (1 - matches["p1 opp service game win rate"])
matches["p2 relative return games Oracle"] = matches["p2 return game win rate Oracle"] - (1 - matches["p2 opp service game win rate"])

matches["p1 relative set Oracle"] = matches["p1 set win rate Oracle"] - (1 - matches["p1 opp set win rate"])
matches["p2 relative set Oracle"] = matches["p2 set win rate Oracle"] - (1 - matches["p2 opp set win rate"])

matches["p1 relative tiebreak Oracle"] = matches["p1 tiebreak win rate Oracle"] - (1 - matches["p1 opp tiebreak win rate"])
matches["p2 relative tiebreak Oracle"] = matches["p2 tiebreak win rate Oracle"] - (1 - matches["p2 opp tiebreak win rate"])
```

```python
# Computing player-tournament stats
matches["p1 tourney serve adv Oracle"] = (matches["tourney serve win rate"] + matches["p1 relative serve Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p1 tourney return adv Oracle"] = (matches["tourney return win rate"] + matches["p1 relative return Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p2 tourney serve adv Oracle"] = (matches["tourney serve win rate"] + matches["p2 relative serve Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p2 tourney return adv Oracle"] = (matches["tourney return win rate"] + matches["p2 relative return Oracle"]).clip(upper=0.9985, lower=0.0015)

matches["p1 tourney service game adv Oracle"] = (matches["tourney service game win rate"] + matches["p1 relative service games Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p1 tourney return game adv Oracle"] = (matches["tourney return game win rate"] + matches["p1 relative return games Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p2 tourney service game adv Oracle"] = (matches["tourney service game win rate"] + matches["p2 relative service games Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p2 tourney return game adv Oracle"] = (matches["tourney return game win rate"] + matches["p2 relative return games Oracle"]).clip(upper=0.9985, lower=0.0015)

matches["p1 tourney set adv Oracle"] = (0.5 + matches["p1 relative set Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p2 tourney set adv Oracle"] = (0.5 + matches["p2 relative set Oracle"]).clip(upper=0.9985, lower=0.0015)

matches["p1 tourney tiebreak adv Oracle"] = (0.5 + matches["p1 relative tiebreak Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p2 tourney tiebreak adv Oracle"] = (0.5 + matches["p2 relative tiebreak Oracle"]).clip(upper=0.9985, lower=0.0015)
```

```python
# Computing matchup statistics
matchup_cols = ["p1 serve_tt", "p1 service_game_tt", "p1 set_tt", "p1 tiebreak_tt", "p2 serve_tt", "p2 service_game_tt"]
matches[matchup_cols] = matches.apply(lambda match: compute_matchup_stats_Oracle(match), axis=1, result_type="expand")
```

### Compute Oracle Shrunken rates using coefs2

```python
matches["p1 serve win rate Oracle"] = matches["p1 serve win rate"] - matches["p1 oracle coef serve2"] * (matches["p1 serve win rate"] - matches["all player year serve win rate"])
matches["p1 return win rate Oracle"] = matches["p1 return win rate"] - matches["p1 oracle coef return2"] * (matches["p1 return win rate"] - (1 - matches["all player year serve win rate"]))
matches["p1 service game win rate Oracle"] = matches["p1 service game win rate"] - matches["p1 oracle coef service game2"] * (matches["p1 service game win rate"] - matches["all player year service game win rate"])
matches["p1 return game win rate Oracle"] = matches["p1 return game win rate"] - matches["p1 oracle coef return game2"] * (matches["p1 return game win rate"] - (1 - matches["all player year service game win rate"]))
matches["p1 set win rate Oracle"] = matches["p1 set win rate"] - matches["p1 oracle coef set2"] * (matches["p1 set win rate"] - 0.5)
matches["p1 tiebreak win rate Oracle"] = matches["p1 tiebreak win rate"] - matches["p1 oracle coef tiebreak2"] * (matches["p1 tiebreak win rate"] - 0.5)

matches["p2 serve win rate Oracle"] = matches["p2 serve win rate"] - matches["p2 oracle coef serve2"] * (matches["p2 serve win rate"] - matches["all player year serve win rate"])
matches["p2 return win rate Oracle"] = matches["p2 return win rate"] - matches["p2 oracle coef return2"] * (matches["p2 return win rate"] - (1 - matches["all player year serve win rate"]))
matches["p2 service game win rate Oracle"] = matches["p2 service game win rate"] - matches["p2 oracle coef service game2"] * (matches["p2 service game win rate"] - matches["all player year service game win rate"])
matches["p2 return game win rate Oracle"] = matches["p2 return game win rate"] - matches["p2 oracle coef return game2"] * (matches["p2 return game win rate"] - (1 - matches["all player year service game win rate"]))
matches["p2 set win rate Oracle"] = matches["p2 set win rate"] - matches["p2 oracle coef set2"] * (matches["p2 set win rate"] - 0.5)
matches["p2 tiebreak win rate Oracle"] = matches["p2 tiebreak win rate"] - matches["p2 oracle coef tiebreak2"] * (matches["p2 tiebreak win rate"] - 0.5)
```

```python
# calculate no sos adjust p1 serve_tt, pl_service_game_tt, etc.
matches["p1 serve_tt Oracle no sos adjust"] = (matches["p1 serve win rate Oracle"] * (1 - matches["p2 return win rate Oracle"])) / ((matches["p1 serve win rate Oracle"] * (1 - matches["p2 return win rate Oracle"])) + (matches["p2 return win rate Oracle"] * (1 - matches["p1 serve win rate Oracle"])))
matches["p2 serve_tt Oracle no sos adjust"] = (matches["p2 serve win rate Oracle"] * (1 - matches["p1 return win rate Oracle"])) / ((matches["p2 serve win rate Oracle"] * (1 - matches["p1 return win rate Oracle"])) + (matches["p1 return win rate Oracle"] * (1 - matches["p2 serve win rate Oracle"])))

matches["p1 service_game_tt Oracle no sos adjust"] = (matches["p1 service game win rate Oracle"] * (1 - matches["p2 return game win rate Oracle"])) / ((matches["p1 service game win rate Oracle"] * (1 - matches["p2 return game win rate Oracle"])) + (matches["p2 return game win rate Oracle"] * (1 - matches["p1 service game win rate Oracle"])))
matches["p2 service_game_tt Oracle no sos adjust"] = (matches["p2 service game win rate Oracle"] * (1 - matches["p1 return game win rate Oracle"])) / ((matches["p2 service game win rate Oracle"] * (1 - matches["p1 return game win rate Oracle"])) + (matches["p1 return game win rate Oracle"] * (1 - matches["p2 service game win rate Oracle"])))

matches["p1 tiebreak_tt Oracle no sos adjust"] = (matches["p1 tiebreak win rate Oracle"] * (1 - matches["p2 tiebreak win rate Oracle"])) / ((matches["p1 tiebreak win rate Oracle"] * (1 - matches["p2 tiebreak win rate Oracle"])) + (matches["p2 tiebreak win rate Oracle"] * (1 - matches["p1 tiebreak win rate Oracle"])))

matches["p1 set_tt Oracle no sos adjust"] = (matches["p1 set win rate Oracle"] * (1 - matches["p2 set win rate Oracle"])) / ((matches["p1 set win rate Oracle"] * (1 - matches["p2 set win rate Oracle"])) + (matches["p2 set win rate Oracle"] * (1 - matches["p1 set win rate Oracle"])))


```

```python
# Generate relative abilities JS, we use unshrunken opp stats (allows a possibility for improvement)
matches["p1 relative serve Oracle"] = matches["p1 serve win rate Oracle"] - (1 - matches["p1 opp return win rate"])
matches["p2 relative serve Oracle"] = matches["p2 serve win rate Oracle"] - (1 - matches["p2 opp return win rate"])

matches["p1 relative return Oracle"] = matches["p1 return win rate Oracle"] - (1 - matches["p1 opp serve win rate"])
matches["p2 relative return Oracle"] = matches["p2 return win rate Oracle"] - (1 - matches["p2 opp serve win rate"])

matches["p1 relative service games Oracle"] = matches["p1 service game win rate Oracle"] - (1 - matches["p1 opp return game win rate"])
matches["p2 relative service games Oracle"] = matches["p2 service game win rate Oracle"] - (1 - matches["p2 opp return game win rate"])

matches["p1 relative return games Oracle"] = matches["p1 return game win rate Oracle"] - (1 - matches["p1 opp service game win rate"])
matches["p2 relative return games Oracle"] = matches["p2 return game win rate Oracle"] - (1 - matches["p2 opp service game win rate"])

matches["p1 relative set Oracle"] = matches["p1 set win rate Oracle"] - (1 - matches["p1 opp set win rate"])
matches["p2 relative set Oracle"] = matches["p2 set win rate Oracle"] - (1 - matches["p2 opp set win rate"])

matches["p1 relative tiebreak Oracle"] = matches["p1 tiebreak win rate Oracle"] - (1 - matches["p1 opp tiebreak win rate"])
matches["p2 relative tiebreak Oracle"] = matches["p2 tiebreak win rate Oracle"] - (1 - matches["p2 opp tiebreak win rate"])
```

```python
# Computing player-tournament stats
matches["p1 tourney serve adv Oracle"] = (matches["tourney serve win rate"] + matches["p1 relative serve Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p1 tourney return adv Oracle"] = (matches["tourney return win rate"] + matches["p1 relative return Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p2 tourney serve adv Oracle"] = (matches["tourney serve win rate"] + matches["p2 relative serve Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p2 tourney return adv Oracle"] = (matches["tourney return win rate"] + matches["p2 relative return Oracle"]).clip(upper=0.9985, lower=0.0015)

matches["p1 tourney service game adv Oracle"] = (matches["tourney service game win rate"] + matches["p1 relative service games Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p1 tourney return game adv Oracle"] = (matches["tourney return game win rate"] + matches["p1 relative return games Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p2 tourney service game adv Oracle"] = (matches["tourney service game win rate"] + matches["p2 relative service games Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p2 tourney return game adv Oracle"] = (matches["tourney return game win rate"] + matches["p2 relative return games Oracle"]).clip(upper=0.9985, lower=0.0015)

matches["p1 tourney set adv Oracle"] = (0.5 + matches["p1 relative set Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p2 tourney set adv Oracle"] = (0.5 + matches["p2 relative set Oracle"]).clip(upper=0.9985, lower=0.0015)

matches["p1 tourney tiebreak adv Oracle"] = (0.5 + matches["p1 relative tiebreak Oracle"]).clip(upper=0.9985, lower=0.0015)
matches["p2 tourney tiebreak adv Oracle"] = (0.5 + matches["p2 relative tiebreak Oracle"]).clip(upper=0.9985, lower=0.0015)
```
