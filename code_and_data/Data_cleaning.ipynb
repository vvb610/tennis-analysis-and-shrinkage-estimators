```python
def analyze_match(match):
    points = match["pbp"]
    points += "."
    data = {}
    
    # server1, server2
    servers = [1, -1]
    # serves won, serves total, service games won, service games total, sets won, tiebreaks won
    metrics = ["sw", "st", "sgw", "sgt", "setsw", "tw"]
    
    for server in servers:
        for metric in metrics:
            data[(server, metric)] = 0
    
    p = 0
    server = 1
    last_point_winner = 0
    while p < len(points):
        if points[p] == "S" or points[p] == "A":
            # server won the point
            data[(server, "sw")] += 1
            data[(server, "st")] += 1
            last_point_winner = server
        elif points[p] == "R" or points[p] == "D":
            # returner won the point
            data[(server, "st")] += 1
            last_point_winner = server * -1
        elif points[p] == ";":
            # end of the game
            if last_point_winner == server:
                data[(server, "sgw")] += 1
            data[(server, "sgt")] += 1
            
            server *= -1
        elif points[p] == ".":
            # set finished, not in tiebreak
            if last_point_winner == server:
                data[(server, "setsw")] += 1
                data[(server, "sgw")] += 1
            else:
                data[(server * -1, "setsw")] += 1
            data[(server, "sgt")] += 1
            
            server *= -1
        elif points[p] == "/":
            # Entered tiebreak
            next_set_server = server * -1
            while p < len(points) and points[p] != ".":
                if points[p] == "/":
                    server *= -1
                elif points[p] == "S" or points[p] == "A":
                    # server won the point
                    data[(server, "sw")] += 1
                    data[(server, "st")] += 1
                    last_point_winner = server
                elif points[p] == "R" or points[p] == "D":
                    data[(server, "st")] += 1
                    last_point_winner = server * -1
                p += 1
            
            data[(last_point_winner, "tw")] += 1
            data[(last_point_winner, "setsw")] += 1
            server = next_set_server
        p += 1
        
    
    return data[(1, "sw")], data[(1,"st")], data[(
        1,"sgw")], data[(1,"sgt")], data[(1,"setsw")], data[(1,"tw")], data[(-1,"sw")], data[(
        -1,"st")], data[(-1,"sgw")], data[(-1,"sgt")], data[(-1,"setsw")], data[(-1,"tw")]
    
```

```python
def compute_tourney_level_stats(match, matches):
    date = match["date"]
    tny_name = match["tny_name"]
    
    one_year_prior = date - relativedelta(years=1)
    relevant_matches = matches[(matches["date"] < date) & (
        matches["date"] >= one_year_prior) & (
        matches["tny_name"] == tny_name)]
    
    relevant_matches["serves won"] = relevant_matches["p1 serves won"] + relevant_matches["p2 serves won"]
    relevant_matches["serves total"] = relevant_matches["p1 serves total"] + relevant_matches["p2 serves total"]
    relevant_matches["service games won"] = relevant_matches["p1 service games won"] + relevant_matches["p2 service games won"]
    relevant_matches["service games total"] = relevant_matches["p1 service games total"] + relevant_matches["p2 service games total"]
    
    aggs = {'serves won':'sum','serves total':'sum', 'service games won':'sum','service games total':'sum'}
    tourney_stats = relevant_matches.groupby(['tny_name']).agg(aggs)
    
    if tourney_stats.shape[0] == 1:
        tourney_serves_won = tourney_stats.iloc[0]["serves won"]
        tourney_serves_total = tourney_stats.iloc[0]["serves total"]
        tourney_service_games_won = tourney_stats.iloc[0]["service games won"]
        tourney_service_games_total = tourney_stats.iloc[0]["service games total"]
    else:
        tourney_serves_won = float("nan")
        tourney_serves_total = float("nan")
        tourney_service_games_won = float("nan")
        tourney_service_games_total = float("nan")
    
    return tourney_serves_won, tourney_serves_total, tourney_service_games_won, tourney_service_games_total
```

```python
def compute_players_stats_all_matches(match, matches):
    p1 = match["server1"]
    p2 = match["server2"]
    
    serve_count = {p1:[0,0], p2:[0,0]}
    return_count = {p1:[0,0], p2:[0,0]}
    service_game_count = {p1:[1,2], p2:[1,2]}
    return_game_count = {p1:[1,2], p2:[1,2]}
    sets_count = {p1:[1,2], p2:[1,2]}
    tiebreaks_count = {p1:[1,2], p2:[1,2]}
    match_results = {p1: "", p2: ""}
    match_count = {p1: [0,0], p2: [0,0]}
    
    all_player_count = {}
    all_player_count["serves won"] = 0
    all_player_count["serves total"] = 0
    all_player_count["service games won"] = 0
    all_player_count["service games total"] = 0

    for index, match_ in matches.iterrows():
        player1 = match_["server1"]
        player2 = match_["server2"]
        
        if player1 not in serve_count:
            serve_count[player1] = [0,0]
            return_count[player1] = [0,0]
            service_game_count[player1] = [1,2]
            return_game_count[player1] = [1,2]
            sets_count[player1] = [1,2]
            tiebreaks_count[player1] = [1,2]
            match_results[player1] = ""
            match_count[player1] = [0,0]
        if player2 not in serve_count:
            serve_count[player2] = [0,0]
            return_count[player2] = [0,0]
            service_game_count[player2] = [1,2]
            return_game_count[player2] = [1,2]
            sets_count[player2] = [1,2]
            tiebreaks_count[player2] = [1,2]
            match_results[player2] = ""
            match_count[player2] = [0,0]
        
        serve_count[player1][0] += match_["p1 serves won"]
        serve_count[player2][0] += match_["p2 serves won"]
        serve_count[player1][1] += match_["p1 serves total"]
        serve_count[player2][1] += match_["p2 serves total"]
        all_player_count["serves won"] += match_["p1 serves won"] + match_["p2 serves won"]
        all_player_count["serves total"] += match_["p1 serves total"] + match_["p2 serves total"]


        return_count[player1][0] += match_["p2 serves total"] - match_["p2 serves won"]
        return_count[player2][0] += match_["p1 serves total"] - match_["p1 serves won"]
        return_count[player1][1] += match_["p2 serves total"]
        return_count[player2][1] += match_["p1 serves total"]

        
        service_game_count[player1][0] += match_["p1 service games won"]
        service_game_count[player2][0] += match_["p2 service games won"]
        service_game_count[player1][1] += match_["p1 service games total"]
        service_game_count[player2][1] += match_["p2 service games total"]
        all_player_count["service games won"] += match_["p1 service games won"] + match_["p2 service games won"]
        all_player_count["service games total"] += match_["p1 service games total"] + match_["p2 service games total"]

        
        return_game_count[player1][0] += match_["p2 service games total"] - match_["p2 service games won"]
        return_game_count[player2][0] += match_["p1 service games total"] - match_["p1 service games won"]
        return_game_count[player1][1] += match_["p2 service games total"]
        return_game_count[player2][1] += match_["p1 service games total"]

        
        sets_count[player1][0] += match_["p1 sets won"]
        sets_count[player2][0] += match_["p2 sets won"]
        sets_count[player1][1] += match_["sets total"]
        sets_count[player2][1] += match_["sets total"]

        
        tiebreaks_count[player1][0] += match_["p1 tiebreaks won"]
        tiebreaks_count[player2][0] += match_["p2 tiebreaks won"]
        tiebreaks_count[player1][1] += match_["p1 tiebreaks won"] + match_["p2 tiebreaks won"]
        tiebreaks_count[player2][1] += match_["p1 tiebreaks won"] + match_["p2 tiebreaks won"]

        if match_["winner"] == 1:
            match_count[player1][0] += 1
            match_count[player1][1] += 1
            match_count[player2][1] += 1
            
            match_results[player1] += "W"
            match_results[player2] += "L"
        else:
            match_count[player2][0] += 1
            match_count[player2][1] += 1
            match_count[player1][1] += 1
            
            match_results[player1] += "L"
            match_results[player2] += "W"
    
    
    return serve_count, return_count, service_game_count, return_game_count, sets_count, tiebreaks_count, match_count, match_results, all_player_count
    
```

```python
def compute_opp_stats_for_match(match, matches, sc, rc, sgc, rgc, setsc, tc, mc):
    p1 = match["server1"]
    p2 = match["server2"]
    players = {p1, p2}
    
    # serve_count_opp, return_count_opp
    sc_opp = {p1:0, p2:0}
    rc_opp = {p1:0, p2:0}
    
    # service_game_count_opp, return_game_count_opp
    sgc_opp = {p1:0, p2:0}
    rgc_opp = {p1:0, p2:0}
    
    # sets_count_opp, tiebreak_count_opp
    setsc_opp = {p1:0, p2:0}
    tc_opp = {p1:0, p2:0}
    mc_opp = {p1:0, p2: 0}
    
    for index, match_ in matches.iterrows():
        player1 = match_["server1"]
        player2 = match_["server2"]
        if player1 not in players and player2 not in players:
            continue
        
        if player1 not in sc_opp:
            sc_opp[player1] = 0
            rc_opp[player1] = 0
            sgc_opp[player1] = 0
            rgc_opp[player1] = 0
            setsc_opp[player1] = 0
            tc_opp[player1] = 0
            mc_opp[player1] = 0
        if player2 not in sc_opp:
            sc_opp[player2] = 0
            rc_opp[player2] = 0
            sgc_opp[player2] = 0
            rgc_opp[player2] = 0
            setsc_opp[player2] = 0
            tc_opp[player2] = 0
            mc_opp[player2] = 0
        
        if sc[player2][1] != 0:
            sc_opp[player1] += (sc[player2][0] / sc[player2][1]) * match_["p1 serves total"]
        if sc[player1][1] != 0:
            sc_opp[player2] += (sc[player1][0] / sc[player1][1]) * match_["p2 serves total"]
            
        if rc[player2][1] != 0:
            rc_opp[player1] += (rc[player2][0] / rc[player2][1]) * match_["p1 serves total"]
        if rc[player1][1] != 0:
            rc_opp[player2] += (rc[player1][0] / rc[player1][1]) * match_["p2 serves total"]
        
        if sgc[player2][1] != 0:
            sgc_opp[player1] += (sgc[player2][0] / sgc[player2][1]) * match_["p1 service games total"]
        if sgc[player1][1] != 0:
            sgc_opp[player2] += (sgc[player1][0] / sgc[player1][1]) * match_["p2 service games total"]
            
        if rgc[player2][1] != 0:
            rgc_opp[player1] += (rgc[player2][0] / rgc[player2][1]) * match_["p1 service games total"]
        if rgc[player1][1] != 0:
            rgc_opp[player2] += (rgc[player1][0] / rgc[player1][1]) * match_["p2 service games total"]
            
        if setsc[player2][1] != 0:
            setsc_opp[player1] += (setsc[player2][0] / setsc[player2][1]) * match_["sets total"]
        if setsc[player1][1] != 0:
            setsc_opp[player2] += (setsc[player1][0] / setsc[player1][1]) * match_["sets total"]
        
        if tc[player2][1] != 0:
            tc_opp[player1] += (tc[player2][0] / tc[player2][1]) * (match_["p1 tiebreaks won"] + match_["p2 tiebreaks won"])
        if tc[player1][1] != 0:
            tc_opp[player2] += (tc[player1][0] / tc[player1][1]) * (match_["p1 tiebreaks won"] + match_["p2 tiebreaks won"])
            
        if mc[player2][1] != 0:
            mc_opp[player1] += (mc[player2][0] / mc[player2][1]) * (1)
        if mc[player1][1] != 0:
            mc_opp[player2] += (mc[player1][0] / mc[player1][1]) * (1)
        
    for player in [p1, p2]:
        if sc[player][1] != 0:
            sc_opp[player] = sc_opp[player] / sc[player][1]
        else:
            sc_opp[player] = float("nan")
        
        if sc[player][1] != 0 and rc[player][1] != 0:
            rc_opp[player] = rc_opp[player] / rc[player][1]
        else:
            rc_opp[player] = float("nan")
            
        if sc[player][1] != 0 and sgc[player][1] > 2:
            sgc_opp[player] = sgc_opp[player] / sgc[player][1]
        else:
            sgc_opp[player] = float("nan")
        
        if sc[player][1] != 0 and rgc[player][1] > 2:
            rgc_opp[player] = rgc_opp[player] / rgc[player][1]
        else:
            rgc_opp[player] = float("nan")
        
        if sc[player][1] != 0 and setsc[player][1] > 2:
            setsc_opp[player] = setsc_opp[player] / setsc[player][1]
        else:
            setsc_opp[player] = float("nan")
        
        if sc[player][1] != 0 and tc[player][1] > 2:
            tc_opp[player] = tc_opp[player] / tc[player][1]
        else:
            tc_opp[player] = float("nan")
            
        if sc[player][1] != 0 and tc[player][1] > 2:
            mc_opp[player] = mc_opp[player] / mc[player][1]
        else:
            mc_opp[player] = float("nan")
    
    p1_data = {
        "sc_opp": sc_opp[p1],
        "rc_opp": rc_opp[p1],
        "sgc_opp": sgc_opp[p1],
        "rgc_opp": rgc_opp[p1],
        "setsc_opp": setsc_opp[p1],
        "tc_opp": tc_opp[p1],
        "mc_opp": mc_opp[p1]
    }
    p2_data = {
        "sc_opp": sc_opp[p2],
        "rc_opp": rc_opp[p2],
        "sgc_opp": sgc_opp[p2],
        "rgc_opp": rgc_opp[p2],
        "setsc_opp": setsc_opp[p2],
        "tc_opp": tc_opp[p2],
        "mc_opp": mc_opp[p2]
    }

    return p1_data, p2_data
            
    
    
```

```python
def compute_matchup_stats(match):
    s_itt = match["p1 tourney serve adv"]
    r_itt = match["p1 tourney return adv"]
    
    s_jtt = match["p2 tourney serve adv"]
    r_jtt = match["p2 tourney return adv"]
    
    sg_itt = match["p1 tourney service game adv"]
    rg_itt = match["p1 tourney return game adv"]
    
    sg_jtt = match["p2 tourney service game adv"]
    rg_jtt = match["p2 tourney return game adv"]
    
    sets_itt = match["p1 tourney set adv"]
    sets_jtt = match["p2 tourney set adv"]
    
    t_itt = match["p1 tourney tiebreak adv"]
    t_jtt = match["p2 tourney tiebreak adv"]
    
    # Pr[i wins serve on j at tournament t on day t]
    s_ijtt = (s_itt * (1 - r_jtt)) / ((s_itt * (1 - r_jtt)) + (r_jtt * (1 - s_itt)))
    # Pr[j wins serve on i at tournament t on day t]
    s_jitt = (s_jtt * (1 - r_itt)) / ((s_jtt * (1 - r_itt)) + (r_itt * (1 - s_jtt)))
    
    # Pr[i wins service game on j at tournament t on day t]
    sg_ijtt = (sg_itt * (1 - rg_jtt)) / ((sg_itt * (1 - rg_jtt)) + (rg_jtt * (1 - sg_itt)))
    # Pr[j wins service game on i at tournament t on day t]
    sg_jitt = (sg_jtt * (1 - rg_itt)) / ((sg_jtt * (1 - rg_itt)) + (rg_itt * (1 - sg_jtt)))
    
    # Pr[i wins set on j at tournament t on day t]
    sets_ijtt = (sets_itt * (1 - sets_jtt)) / ((sets_itt * (1 - sets_jtt)) + (sets_jtt * (1 - sets_itt)))
    
    # Pr[i wins tiebreak on j at tournament t on day t]
    t_ijtt = (t_itt * (1 - t_jtt)) / ((t_itt * (1 - t_jtt)) + (t_jtt * (1 - t_itt)))
    
    
    return s_ijtt, sg_ijtt, sets_ijtt, t_ijtt, s_jitt, sg_jitt
    
```

```python
def calculate_shrunken_values(match, variance_type, matches):
    # determine value fields
    p1_value_cols = ["p1 serve win rate", "p1 return win rate", "p1 service game win rate", "p1 return game win rate", "p1 set win rate", "p1 tiebreak win rate"]
    p2_value_cols = ["p2 serve win rate", "p2 return win rate", "p2 service game win rate", "p2 return game win rate", "p2 set win rate", "p2 tiebreak win rate"]
    value_cols = p1_value_cols + p2_value_cols
    
    # determine opp/ T cols
    p1_opp_cols = ["p1 opp serve win rate", "p1 opp return win rate", "p1 opp service game win rate", "p1 opp return game win rate", "p1 opp set win rate", "p1 opp tiebreak win rate"]
    p2_opp_cols = ["p2 opp serve win rate", "p2 opp return win rate", "p2 opp service game win rate", "p2 opp return game win rate", "p2 opp set win rate", "p2 opp tiebreak win rate"]
    opp_cols = p1_opp_cols + p2_opp_cols
    
    # determine variance fields
    p1_cols_variances = ["p1 serve win rate variance", "p1 return win rate variance", "p1 service game win rate variance", "p1 return game win rate variance", "p1 set win rate variance", "p1 tiebreak win rate variance"]
    p2_cols_variances = ["p2 serve win rate variance", "p2 return win rate variance", "p2 service game win rate variance", "p2 return game win rate variance", "p2 set win rate variance", "p2 tiebreak win rate variance"]
    
    if variance_type != 1:
        for i in range(len(p1_cols_variances)):
            p1_cols_variances[i] += str(variance_type)
            p2_cols_variances[i] += str(variance_type)
            
    variance_cols = p1_cols_variances + p2_cols_variances
    
    shrunk_vals = []
    for i in range(len(value_cols)):
        temp = match[value_cols[i]] - ((match[variance_cols[i]]) / (match[variance_cols[i]] + (match[value_cols[i]] - match[opp_cols[i]])**2)) * (match[value_cols[i]] - match[opp_cols[i]])
        shrunk_vals.append(temp)
        
    # maybe we want to recalculate opp values but this will involve a more intensive refactor
    
    return tuple(shrunk_vals)
```

## Defining opponent-dependent stats

For game $(i,k,t)$ between players $i,k$ happening at time $t$, define:

> $SOSS_{it} := \frac{1}{total\ points\ i\ returned}\sum_{j \in opp(i)} \frac{points\ won\ on\ serve\ by j}{total\ points\ served\ by\ j} * (points\ i\ returned\ on\ j)$

> $SOSR_{it} := \frac{1}{total\ points\ i\ served}\sum_{j \in opp(i)} \frac{points\ won\ on\ return\ by j}{total\ points\ returned\ by\ j} * (points\ i\ served\ on\ j)$

Where we restrict dataset to the 1 year period before and up to $t$

$SOSS_{it}$ intends to capture the strength of servers $i$ faced in the past year prior to $t$.

$SOSR_{it}$ intends to capture the strength of returners $i$ faced in the past year prior to $t$.

We can define similar statistics to measure the strength of service games, return games, sets, and tiebreaks $i$ faced. These stats demonstrate different granularities in difficulty of the opponents level.

## Computing matchup stats at time $t$

#### Sidebar- tense match up between $i$ and $j$

Let's take a sidebar. Suppose it's the start of a tennis season. There are two tennis players $(i,j)$ and they both talk smack that they will be the better player at the end of the season and schedule a match. For sake of argument, suppose that the tennis season consists of a league of N players and they all play each other twice so that they face the same strength of schedule. The tennis smack talk becomes so contentious that at the end of the season, it's unsafe for the two to play. Nonetheless they want to know who is the better player. Player $k$ jumps in, he was an average player with a winning percentage of $50\%$ over the season, and proposes that both $i$ and $j$ play versus $k$ with the following rules:

> If $i$ beats $k$ and $k$ beats $j$, $i$ is declared the better player.

> If $j$ beats $k$ and $k$ beats $i$, $j$ is declared the better player.

> If ($i$ beats $k$ and $j$ beats $k$) or ($k$ beats $i$ and $k$ beats $j$), they do the matches again.

Suppose all matches are independent and player $i$ posed a winning percentage of $WP_i$ and player $j$ posed a winning percentage of $WP_j$ over the long season. Then, since $k$ has a winning percentage of $50\%$, that means:

> $\Pr$\[$i$ beats $k$\] = $WP_i$

> $\Pr$\[$j$ beats $k$\] = $WP_j$

since player $i$ posed a winning percentage of $WP_i$ versus a group that has an average winning percentage of $50\%$* and player $k$ has a winning percentage of $50\%$. Vis-a-vis for player $j$. In other words, while $WP_i$ is a measure of the percentage of games $i$ won over the season, it's also a measure of how likely $i$ is to beat an average player in the league.

As a result, we can say

> $\Pr$\[$i$ declared better player\] $= \frac{(WP_i)(1 - WP_j)}{(WP_i)(1 - WP_j) + (1 - WP_i)(WP_j)}$


*Technically $i$ probably didn't face an average winning percentage of $50\%$ over the season. It's only true that $\frac{1}{N}\sum_{l=0}^N WP_l = 50\%$. In fact $i$ faced an average winning percentage of $\frac{1}{N-1}\sum_{l=0, l \neq i}^N WP_l$ but for large enough $N$, this is a reasonable approximation. 

#### Computing matchup $(i,j,t)$ stats

For game $(i,j,t)$ between players $i,j$ happening at time $t$, among other quantities, we want to figure out the probability that player $i$ will win a serve on player $j$. 

In our previous data analysis, we have sought to correct $i$'s serve ability for strength of returners faced by player $i$. 

Recall that we defined:

> $SOSR_{it} := \frac{1}{total\ points\ i\ served}\sum_{j \in opp(i)} \frac{points\ won\ on\ return\ by j}{total\ points\ returned\ by\ j} * (points\ i\ served\ on\ j)$

> $opp(i) := $ $i$'s opponents in the last year

This is a weighted average over $i$'s opponents over the past year (weighted by number of points $i$ served to each $j$), of $j$'s return win rate. It equals the average return win rate $i$ faced in each of its serve points in the past year. In other words, we can say that the average serve win rate, in the past year, over this set of opponents, is $1 - SOSR_{it}$. 

We also have access to $i$'s serve win rate in the past year, $s_{it}$. That is, 

> $s_{it} := \frac{points\ i\ won\ on\ serve}{total\ points\ i\ served}$. 

We can the define how much better/ worse $i$ served, than an average server against $i$'s opponents in the past year:

> $srel_{it} := s_{it} - (1 - SOSR_{it})$

Folowing an analogous trajectory define how much better/ worse $j$ returned, than an average returner against $j$'s opponents in the past year:

> $rrel_{jt} := r_{jt} - (1 - SOSS_{jt})$

We now have relative serve ability of $i$: (ie., how much more likely $i$ is to win a serve against $opp(i)$ than an average server against this group) and relative return ability of $j$. I propose benchmarking the serve value to $s_tt$, the average serve win rate for all matches across the past year at the tournament. I propose benchmarking the return value to $r_ttt$, the average return win rate for all matches across the past year at the tournament. 

We define:

> $s_{itt} := s_tt + srel_{it}$. 

> $r_{jtt} := r_tt + rrell_{jt}$

What we have constructed in $s_{itt}$, is an opponent-and-tournament-corrected probability that I will win a serve on day $t$. In other words, I declare $s_{itt}$ an estimate of the probability that $i$ will win a serve against an average opponent at tournament $t$ on day $t$. Vis-a-vis for $r_{jtt}$.

If we think back to the sidebar, $s_{itt}$ and $r_{jtt}$ look analogous to $WP_i$ and $WP_j$ since they are all equal to the probability of "success" against an average opponenet (where success is aptly defined in each case).

As a result, I define

> $s_{ijtt} := \Pr$\[$i$ wins serve against $j$ on day $t$ at tournament $t$\] = $\frac{(s_{itt})(1 - r_{jtt})}{(s_{itt})(1 - r_{jtt}) + (r_{jtt})(1 - s_{itt})}$

as the probability that $i$ beats $j$ on a serve on day $t$ at tournament $t$. Consequently,

> $r_{jitt} := 1 - s_{ijtt}$.

We can follow similar trajectories to calculate the probability that $j$ wins a serve against $i$, $i$ wins a service game against $j$, $j$ wins a service game against $i$, $i$ wins a set against $j$, and $i$ wins a tiebreak against $j$. The other probabilities yield from these quantites and the law of total probability.

# Introducing Shrinkage Estimators: Oracle and James-Stein

### Oracle Estimator

For simplicity, I will focus my analysis thinking about the probability a player wins a point on a serve, but it's applicable to all other player statistics we were computing above (ie., return ability, service game win ability, etc.)

Consider a game between p1 and p2 in which p1 wins $s_{12}$ fraction of his $N_1$ points on serve, in accordance with the data we've collected. We continue to assume all of p1's serves are iid. 

We want to estimate $\sigma$, the underlying probability that p1 wins a serve on p2. We observe $s$, the fraction of points won on serve by p1 and that $s$ has variance $v^2$. We'll call our estimate $\hat{\sigma}$. 

$MLE(\sigma) = argmax_{s \in [0,1]} \Pr(s_{12} | s) = s_{12}$. 

However, another criteria to estimate $\sigma$, is to minimize the mean squared error. 

$MSE(\sigma) = \mathbb{E}[(\hat{\sigma} - \sigma)^2] = (\sigma - \mathbb{E}[\hat{\sigma}])^2 + \mathbb{E}[(\mathbb{E}[\hat{\sigma}] - \hat{\sigma})^2]$ = $Bias(\hat{\sigma})^2 + Var(\hat{\sigma})$

$MLE(\sigma)$ is unbiased and therefore the first term drops off. However, it's possible to reduce the mean squared error by introducing some bias.

Define an estimator:

$\hat{\sigma} := s - \gamma(s - T)$ for $\gamma \in [0,1]$ and any $T$. Note that the estimator is biased if $\gamma \neq 0$ and $T \neq s$. 

$MSE[\hat{\sigma}] = $

$= \mathbb{E}[(s - \gamma(s - T) - \sigma)^2] $

$= \mathbb{E}[((s - \sigma) - \gamma(s - T))^2] $

$= \mathbb{E}[(s - \sigma)^2 + \gamma^2(s - T)^2 - 2(s - \sigma)\gamma(s-T)]$

$=v^2 + \gamma^2\mathbb{E}[s^2 + T^2 - 2Ts] - 2\gamma\mathbb{E}[s^2 - sT + T\sigma - s\sigma]$

$= v^2 + \gamma^2((v^2 + \sigma^2) + T^2 - 2T-\sigma) - 2\gamma((v^2 + \sigma^2) - \sigma^2)$

$= v^2 + \gamma^2(v^2 + (T - \sigma)^2)) - 2\gamma v^2$ (*)

Taking 

$\frac{d\ MSE(\hat{\sigma}_o)}{d\gamma} = 2\gamma(v^2 + (T - \sigma)^2) - 2v^2$

Setting that derivative equal to 0 and solving for $\gamma$ gives

$\gamma = \frac{v^2}{v^2 + (T - \sigma)^2}$ and $\hat{\sigma} = s - \frac{v^2}{v^2 + (T - \sigma)^2}(s - T)$

Plugging back into (*):

We get that the optimal $\gamma$ satisfies:

$MSE[\hat{\sigma}] = v^2 - \frac{v^2}{v^2 + (T - \sigma)^2}v^2$, which is always less than $v^2$ no matter the choice of $T$. In fact, it behaves really well when $T$ is close to $\sigma$ as we're able to reduce our mean squared error tremendously. In fact, if you know $\sigma$ beforehand, you can set $T = \sigma$ and get zero mean-squared error. 

The difficulty in directly applying this equation in our case, is that we don't $\sigma$, and therefore we also don't know $v$. That means we don't know what to plug-in for these values when evaluating $\gamma$. Moreover, how should we pick $T$. 

I propose the following problem.

```python
def compute_players_stats_all_matches(matches):

    serve_count = {}
    return_count = {}
    service_game_count = {}
    return_game_count = {}
    sets_count = {}
    tiebreaks_count = {}
    match_count = {}
    
    all_player_count = {}
    all_player_count["serves won"] = 0
    all_player_count["serves total"] = 0
    all_player_count["service games won"] = 0
    all_player_count["service games total"] = 0

    for index, match_ in matches.iterrows():
        player1 = match_["server1"]
        player2 = match_["server2"]
        
        if player1 not in serve_count:
            serve_count[player1] = [0,0,0]
            return_count[player1] = [0,0,0]
            service_game_count[player1] = [0,0,0]
            return_game_count[player1] = [0,0,0]
            sets_count[player1] = [0,0,0]
            tiebreaks_count[player1] = [0,0,0]
            match_count[player1] = 0
        if player2 not in serve_count:
            serve_count[player2] = [0,0,0]
            return_count[player2] = [0,0,0]
            service_game_count[player2] = [0,0,0]
            return_game_count[player2] = [0,0,0]
            sets_count[player2] = [0,0,0]
            tiebreaks_count[player2] = [0,0,0]
            match_count[player2] = 0
        
        serve_count[player1][0] += match_["p1 serves won"]
        serve_count[player2][0] += match_["p2 serves won"]
        serve_count[player1][1] += match_["p1 serves total"]
        serve_count[player2][1] += match_["p2 serves total"]
        all_player_count["serves won"] += match_["p1 serves won"] + match_["p2 serves won"]
        all_player_count["serves total"] += match_["p1 serves total"] + match_["p2 serves total"]
        s_1match = (match_["p1 serves won"] + 1) / (match_["p1 serves total"] + 2)
        s_2match = (match_["p2 serves won"] + 1) / (match_["p2 serves total"] + 2)
        serve_count[player1][2] += match_["p1 serves total"] * s_1match * (1 - s_1match)
        serve_count[player2][2] += match_["p2 serves total"] * s_2match * (1 - s_2match)

        return_count[player1][0] += match_["p2 serves total"] - match_["p2 serves won"]
        return_count[player2][0] += match_["p1 serves total"] - match_["p1 serves won"]
        return_count[player1][1] += match_["p2 serves total"]
        return_count[player2][1] += match_["p1 serves total"]
        r_1match = (match_["p2 serves total"] - match_["p2 serves won"] + 1) / (match_["p2 serves total"] + 2)
        r_2match = (match_["p1 serves total"] - match_["p1 serves won"] + 1) / (match_["p1 serves total"] + 2)
        return_count[player1][2] += match_["p2 serves total"] * r_1match * (1 - r_1match)
        return_count[player2][2] += match_["p1 serves total"] * r_2match * (1 - r_2match)
        
        service_game_count[player1][0] += match_["p1 service games won"]
        service_game_count[player2][0] += match_["p2 service games won"]
        service_game_count[player1][1] += match_["p1 service games total"]
        service_game_count[player2][1] += match_["p2 service games total"]
        all_player_count["service games won"] += match_["p1 service games won"] + match_["p2 service games won"]
        all_player_count["service games total"] += match_["p1 service games total"] + match_["p2 service games total"]
        sg_1match = (match_["p1 service games won"] + 1) / (match_["p1 service games total"] + 2)
        sg_2match = (match_["p2 service games won"] + 1) / (match_["p2 service games total"] + 2)
        service_game_count[player1][2] += match_["p1 service games total"] * sg_1match * (1 - sg_1match)
        service_game_count[player2][2] += match_["p2 service games total"] * sg_2match * (1 - sg_2match)
        
        return_game_count[player1][0] += match_["p2 service games total"] - match_["p2 service games won"]
        return_game_count[player2][0] += match_["p1 service games total"] - match_["p1 service games won"]
        return_game_count[player1][1] += match_["p2 service games total"]
        return_game_count[player2][1] += match_["p1 service games total"]
        rg_1match = (match_["p2 service games total"] - match_["p2 service games won"] + 1) / (match_["p2 service games total"] + 2)
        rg_2match = (match_["p1 service games total"] - match_["p1 service games won"] + 1) / (match_["p1 service games total"] + 2)
        return_game_count[player1][2] += match_["p2 service games total"] * rg_1match * (1 - rg_1match)
        return_game_count[player2][2] += match_["p1 service games total"] * rg_2match * (1 - rg_2match)
        
        sets_count[player1][0] += match_["p1 sets won"]
        sets_count[player2][0] += match_["p2 sets won"]
        sets_count[player1][1] += match_["sets total"]
        sets_count[player2][1] += match_["sets total"]
        sets_1match = (match_["p1 sets won"] + 1) / (match_["sets total"] + 2)
        sets_2match = (match_["p2 sets won"] + 1) / (match_["sets total"] + 2)
        sets_count[player1][2] += match_["sets total"] * sets_1match * (1 - sets_1match)
        sets_count[player2][2] += match_["sets total"] * sets_2match * (1 - sets_2match)
        
        tiebreaks_count[player1][0] += match_["p1 tiebreaks won"]
        tiebreaks_count[player2][0] += match_["p2 tiebreaks won"]
        tiebreaks_count[player1][1] += match_["p1 tiebreaks won"] + match_["p2 tiebreaks won"]
        tiebreaks_count[player2][1] += match_["p1 tiebreaks won"] + match_["p2 tiebreaks won"]
        tb_1match = (match_["p1 tiebreaks won"] + 1) / (match_["p1 tiebreaks won"] + match_["p2 tiebreaks won"] + 2)
        tb_2match = (match_["p2 tiebreaks won"] + 1) / (match_["p1 tiebreaks won"] + match_["p2 tiebreaks won"] + 2)
        tiebreaks_count[player1][2] += (match_["p1 tiebreaks won"] + match_["p2 tiebreaks won"]) * tb_1match * (1 - tb_1match)
        tiebreaks_count[player2][2] += (match_["p1 tiebreaks won"] + match_["p2 tiebreaks won"]) * tb_2match * (1 - tb_2match)
        
        match_count[player1] += 1
        match_count[player2] += 1
    
    for player in match_count:
        if serve_count[player][1] != 0:
            serve_count[player][2] = serve_count[player][2] / (serve_count[player][1])**2
        else:
            serve_count[player][2] = float("nan")
        if return_count[player][1] != 0:
            return_count[player][2] = return_count[player][2] / (return_count[player][1])**2
        else:
            return_count[player][2] = float("nan")
        if service_game_count[player][1] != 0:
            service_game_count[player][2] = service_game_count[player][2] / (service_game_count[player][1])**2
        else:
            service_game_count[player][2] = float("nan")
        if return_game_count[player][1] != 0:
            return_game_count[player][2] = return_game_count[player][2] / (return_game_count[player][1])**2
        else:
            return_game_count[player][2] = float("nan")
        if sets_count[player][1] != 0:
            sets_count[player][2] = sets_count[player][2] / (sets_count[player][1])**2
        else:
            sets_count[player][2] = 0
        if tiebreaks_count[player][1] != 0:
            tiebreaks_count[player][2] = tiebreaks_count[player][2] / (tiebreaks_count[player][1])**2
        else:
            tiebreaks_count[player][2] = float("nan")
    
    return serve_count, return_count, service_game_count, return_game_count, sets_count, tiebreaks_count, match_count, all_player_count
    
```
